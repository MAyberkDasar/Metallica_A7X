```python
import pandas as pd
import seaborn as sns
```

METALLICA 

```python
metallica = pd.read_csv("metallica_songs.csv")
```

```python
metallica.info()
```

```python
metallica.describe().T
```

```python
metallica.head()
```

```python
metallica.drop(columns= ["spotify_uri", "spotify_id", "track_number"], inplace= True)
```

```python
def formatDuration(ms: int):
    total_second = ms / 1000
    minute = int(total_second // 60)
    second = int(total_second % 60)
    return f"{minute} : {second:02}"
```

```python
metallica["duration"] = metallica["duration_ms"].apply(formatDuration)
```

```python
metallica
```

```python
sns.barplot(x= "popularity", y= "album", data= metallica)
```

```python
sns.boxplot(x= "popularity", y= "album", data= metallica)
```

```python
metallica["year"] = pd.to_datetime(metallica["release_date"], format= "mixed").dt.year

sns.lineplot(x= "year", y= "popularity", data= metallica)
```

```python
sns.scatterplot(x= "tempo", y= "popularity", data= metallica)
```

```python
sns.scatterplot(x= "danceability", y= "valence", hue= "energy", data=metallica)
```
